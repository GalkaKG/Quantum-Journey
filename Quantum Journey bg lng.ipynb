# Квантово пътешествие 

## Какво е квант?

Квантът (множествено число: кванти) е най-малката дискретна единица на дадено явление. Например квант светлина е фотон, а квант електричество е електрон. Квантът идва от латински и означава "количество" или "колко?". Ако нещо е количествено измеримо, то може да бъде измерено.

## Какво е квант в компютрите?
Квантовите изчисления използват природата на субатомните частици за извършване на изчисления, вместо да използват електрически сигнали, както е при класическите изчисления. Квантовите компютри използват кюбити вместо двоични битове. Чрез програмиране на началните условия на кюбита квантовите изчисления могат да решат проблем, когато състоянието на суперпозиция се срине. Изследванията на квантовите компютри са насочени към свързването на по-голям брой кюбити, за да могат да се решават по-големи и по-сложни проблеми.

![Alt text](https://github.com/GalkaKG/Quantum-Journey/blob/main/images/classical_computing_vs_quantum_computing.png?raw=true) _Квантовите компютри използват природата на субатомните частици за извършване на изчисления като алтернатива на електрическите сигнали, използвани в класическите компютри._

Квантовите компютри могат да извършват определени изчисления много по-бързо от класическите компютри. За да намерят отговор на даден проблем, класическите компютри трябва да преминат през всяка възможност една по една. При някои видове проблеми преминаването през всички варианти може да отнеме много време. Квантовите компютри нямат нужда да опитват всеки вариант, а решават отговора почти мигновено.

Някои задачи, които квантовите компютри могат да решат по-бързо от класическите, са факторирането на прости числа и задачата за пътуващия търговец. Когато квантовите компютри демонстрират способността си да решават тези задачи по-бързо от класическите компютри, ще бъде постигнато квантово превъзходство.

Проблемът за пътуващия търговец (TSP) е алгоритмична задача, която има за задача да намери най-краткия маршрут между набор от точки и места, които трябва да бъдат посетени. В условието на задачата точките са градовете, които продавачът може да посети. Целта на търговеца е да запази разходите за пътуване и изминатото разстояние възможно най-ниски. Тук можете да прочетете повече за него на адрес [a link](https://www.geeksforgeeks.org/travelling-salesman-problem-using-dynamic-programming/) .

Както се вижда от изображението, операциите в квантовите изчисления се определят чрез линейна алгебра над Хилбертовото пространство.

### Какво е Хилбертово пространство?

Хилбертово пространство (ХП) е понятие в математиката обобщаващо Евклидовото пространство. Наречено е на Давид Хилберт, който пръв въвежда концепцията за безкрайномерно Евклидово пространство през 1909 г.

Хилбертовото пространство разширява методите на векторната алгебра от двумерната равнина и тримерното пространство към многомерните пространства.

Ако трябва да го дефинираме с по-строги математически термини, Хилбертовото пространство е векторно пространство, в което разстоянията и ъглите могат да бъдат измерени и което е пълно. Тоест за всяка редица от вектори на Коши съществува граница в пространството.

В общия случай ХП е безкрайномерно, линейно, векторно пространство над полето на комплексните числа $ \mathbb{C} $ със скаларно произведение, относно което то е пълно.

Пространство на Хилберт е реално или комплексно векторно пространство, което е пълно и в което __нормата__ се определя от скаларното произведение 
$ \langle \cdot , \cdot \rangle $ посредством формулата:

$ \| x \| = \sqrt {\langle x, x \rangle} $

Това означава, че нормата на вектор $ x $ е квадратният корен от скаларното произведение на $ x $ със себе си.

__Събиране__:

Две Хилбертови пространства $ H_1 $ и $ H_2 $ могат да бъдат комбинирани в едно общо Хилбертово пространство, наричано директна ортогонална сума и обозначавано като:

$ H_1 \oplus  H_2 $

В това ново пространство, всеки вектор се представя като двойка $ (x_1, x_2) $ , където $ x_1 \in H_1 $ и $ x_2 \in H_2 $ .

### А какво е квантово превъзходство?
Квантовото превъзходство е експериментална демонстрация на господството и предимството на квантовия компютър пред класическите компютри чрез извършване на невъзможни досега изчисления с несравнима скорост. За да потвърдят, че е постигнато квантово превъзходство, компютърните специалисти трябва да могат да покажат, че класическият компютър никога не би могъл да реши проблема, като същевременно докажат, че квантовият компютър може да извърши изчислението бързо.

Компютърните учени смятат, че квантовото превъзходство ще доведе до разбиването на алгоритъма на Шор - невъзможно в момента изчисление, което е в основата на повечето съвременни криптографии, както и ще предложи предимства в разработването на лекарства, прогнозите за времето, борсовите сделки и дизайна на материали.

Квантовите изчисления непрекъснато се развиват. Квантовите компютри все още не са достигнали точката, в която могат да покажат превъзходството си над класическите компютри. Това се дължи най-вече на огромното количество квантови битове или кюбити, необходими за извършване на смислени операции на квантовите компютри. С увеличаването на количеството необходими логически врати и броя на кюбитите се увеличава и процентът на грешките. Ако процентът на грешките стане твърде висок, квантовият компютър губи всички предимства, които е имал пред класическия компютър.

За успешното извършване на полезни изчисления - като например определяне на химичните свойства на дадено вещество - ще са необходими няколко милиона кюбита. Понастоящем най-големият проект за квантов компютър е квантовият компютър на IBM, наречен Osprey, който разполага с 433 кюбита.

### Какво представлява постквантовата криптография?
Постквантовата криптография, известна също като квантово криптиране, е разработване на криптографски системи за класически компютри, които могат да предотвратяват атаки, извършвани от квантови компютри.

През 80-те години на миналия век учените изказват предположение, че ако компютрите могат да се възползват от уникалните свойства на квантовата механика, те биха могли да извършват сложни изчисления по-бързо от класическите, двоични компютри. Бързо стана ясно, че квантовият компютър, който се възползва от квантовите свойства като суперпозиция и заплитане, може да извърши някои видове сложни изчисления за няколко часа - нещо, което би отнело на класически компютър няколко години.

През 90-те години на миналия век, след като математикът Питър Шор успешно демонстрира, че теоретичен квантов компютър може лесно да разбие алгоритъма, използван за криптиране с публичен ключ (PKE), криптографите по света започнаха да проучват как би изглеждала една постквантова криптографска система. Към момента на писане на този текст стандартите за постквантово криптиране все още се разработват.

През 2016 г. Националният институт за стандарти и технологии (NIST) започна да търси предложения за алгоритми, които потенциално биха могли да заменят криптирането с публичен ключ, механизмите за капсулиране на ключове (KEM) и цифровите подписи. Математици и програмисти започнаха да експериментират с различни стратегии, които да заменят факторизацията на цели числа и дискретните логаритмични задачи, използвани в криптосистемите Rivest-Shamir-Adleman (RSA), Elliptic Curve Digital Signature Algorithm (ECDSA), Elliptic Curve Diffie-Hellman Key Exchange (ECDH) и Digital Signature Algorithm (DSA).

Експериментите на Google в областта на постквантовата криптография, например, включват свързване на класически алгоритъм с елиптична крива с постквантов алгоритъм. Идеята е, че дори квантовата криптография да се окаже разбиваема, добавянето на алгоритъм с елиптична крива все пак ще осигури известна степен на сигурност.

Други популярни стратегии за квантово устойчиви алгоритми включват използването на решетъчни, кодово-базирани и многовариантни схеми. Към момента на написване на този текст решетъчните схеми изглеждат най-обещаващи, тъй като е изключително трудно да се изчисли най-късият вектор на голяма решетка, когато най-късият вектор е квантов и може да съществува в повече от едно измерение.

Алгоритмите, които днес поддържат криптиране, включително криптографията с публичен ключ, се считат за безопасни за електронната търговия. Въпреки че квантовите изчисления са реални, технологията е скъпа и случаите на използване се коренят в научните и правителствените изследвания. Въпреки това надпреварата между изследователите, които се опитват да намерят работещо постквантово криптиране, и изследователите, които се опитват да разбият RSA и подобни криптосистеми с квантови алгоритми, продължава.

Много експерти смятат, че квантовото превъзходство ще бъде постигнато в рамките на девет или десет години, като тогава RSA и подобни асиметрични алгоритми вече няма да могат да защитават чувствителни данни. Затова NIST агресивно се стреми да създаде стандарт за постквантово криптиране.

## Какво представлява квантовата криптография?
Квантовата криптография е метод за криптиране, който използва естествените свойства на квантовата механика за защита и предаване на данни по начин, който не може да бъде хакнат.

Криптографията е процесът на криптиране и защита на данни, така че само лицето, което притежава правилния секретен ключ, да може да ги декриптира. Квантовата криптография се различава от традиционните криптографски системи по това, че разчита на физиката, а не на математиката, като ключов аспект на модела за сигурност.

Квантовата криптография е система, която е напълно защитена срещу компрометиране без знанието на изпращача или получателя на съобщението. Това означава, че е невъзможно да се копират или преглеждат данни, кодирани в квантово състояние, без да се предупреди изпращачът или получателят. Квантовата криптография трябва да остане сигурна и срещу тези, които използват квантови изчисления.

Квантовата криптография използва отделни частици светлина, или фотони, за предаване на данни по оптичен проводник. Фотоните представляват двоични битове. Сигурността на системата се основава на квантовата механика. Тези защитни свойства включват следното:

- Частиците могат да съществуват на повече от едно място или в повече от едно състояние по едно и също време;
- дадено квантово свойство не може да се наблюдава, без да се промени или наруши; и
- цели частици не могат да бъдат копирани.
Тези свойства правят невъзможно измерването на квантовото състояние на която и да е система, без да се наруши състоянието на тази система.

Фотоните се използват за квантовата криптография, защото предлагат всички необходими качества, които са необходими: Поведението им е добре изучено и те са носители на информация в оптичните кабели. Един от най-известните примери за квантова криптография в момента е квантовото разпределение на ключове (QKD), което осигурява сигурен метод за обмен на ключове.

### Как работи квантовата криптография?
На теория квантовата криптография работи по модел, разработен през 1984 г.

Моделът предполага, че има двама души на име Алис и Боб, които искат да обменят сигурно съобщение. Алиса инициира съобщението, като изпраща на Боб ключ. Ключът е поток от фотони, които се движат в една посока. Всеки фотон представлява един бит данни - или 0, или 1. В допълнение към линейното си движение обаче тези фотони осцилират или вибрират по определен начин.

Така че преди Алиса, изпращачът, да инициира съобщението, фотоните преминават през поляризатор. Поляризаторът е филтър, който позволява на определени фотони да преминават през него с едни и същи вибрации, а на други позволява да преминават в променено състояние на вибрация. Поляризираните състояния могат да бъдат вертикални (1 бит), хоризонтални (0 бит), 45 градуса надясно (1 бит) или 45 градуса наляво (0 бит). Предаването има една от двете поляризации, представляващи един бит, 0 или 1, в двете схеми, които тя използва.

Сега фотоните преминават през оптичното влакно от поляризатора към приемника, Боб. При този процес се използва разделител на лъча, който отчита поляризацията на всеки фотон. При получаването на фотонния ключ Боб не знае правилната поляризация на фотоните, затова една поляризация се избира на случаен принцип. Сега Алиса сравнява какво е използвал Боб, за да поляризира ключа, и след това съобщава на Боб кой поляризатор е използвала, за да изпрати всеки фотон. След това Боб потвърждава дали е използвал правилния поляризатор. След това фотоните, прочетени с грешния поляризатор, се изхвърлят, а останалата последователност се счита за ключ.

Да предположим, че има подслушвач, наречен Ева. Ева се опитва да подслушва и разполага със същите инструменти като Боб. Но Боб има предимството да говори с Алиса, за да потвърди кой тип поляризатор е използван за всеки фотон; Ева няма такова предимство. В крайна сметка Ева извежда крайния ключ неправилно.

Алис и Боб също така ще разберат, ако Ева ги подслушва (😁). Тогава Ева, наблюдавайки потока от фотони, би променила позициите на фотоните, които Алиса и Боб очакват да видят.

### Предимства на квантовата криптография
Ползите от квантовата криптография включват следното:

* Осигурява сигурна комуникация. Вместо трудни за разбиване числа, квантовата криптография се основава на законите на физиката, което е по-сложен и сигурен метод за криптиране.
* Открива подслушване. Ако трета страна се опита да прочете кодираните данни, тогава квантовото състояние се променя, променяйки очаквания резултат за потребителите.
* Предлага множество методи за защита. Използват се множество протоколи за квантова криптография. Някои от тях, като QKD например, могат да се комбинират с класически методи за криптиране, за да се повиши сигурността.

### Ограничения на квантовата криптография
Потенциалните недостатъци и ограничения, които идват с квантовата криптография, включват следното:

* Промени в поляризацията и честотата на грешките. Фотоните могат да променят поляризацията си по време на пренасяне, което потенциално увеличава процента на грешките.
* Обхват. Максималният обхват на квантовата криптография обикновено е около 400-500 km, с изключение на Terra Quantum.
* Разходи. Квантовата криптография обикновено изисква собствена инфраструктура, като се използват оптични линии и ретранслатори.
* Брой дестинации. Не е възможно да се изпращат ключове до две или повече места в квантов канал.

Този метод на криптография все още не е напълно разработен, но вече има **успешни реализации**:

* Университетът в Кеймбридж и Toshiba Corp. създадоха система за QKD с висока скорост на предаване, използвайки протокола за квантова криптография BB84.
* Квантовата мрежа на Агенцията за перспективни изследователски проекти в областта на отбраната, която функционираше от 2002 до 2007 г., представляваше 10-възелна QKD мрежа, разработена от Бостънския университет, Харвардския университет и IBM Research.
* Quantum Xchange стартира първата квантова мрежа в САЩ, включваща 1000 км оптичен кабел.
* Търговски компании, като ID Quantique, Toshiba, Quintessence Labs и MagiQ Technologies Inc. също разработиха търговски QKD системи.

В допълнение към QKD някои от по-значимите **протоколи** и **квантови алгоритми**, използвани в квантовата криптография, са следните:

* квантово хвърляне на монети;
* квантова криптография, базирана на позицията; и
* независима от устройствата квантова криптография.

Един от алгоритмите на квантовите изчисления, алгоритъм за факторизация, който се дължи на Питър Шор, ще бъде на фокус сега.

## Алгоритъм на Шор

Алгоритъмът на Шор е квантов алгоритъм за намиране на простите фактори на цяло число. Той е разработен през 1994 г. от американския математик Питър Шор. Това е един от малкото известни квантови алгоритми с убедителни потенциални приложения и силни доказателства за суперполиномиално ускорение в сравнение с най-известните класически (неквантови) алгоритми. От друга страна, факторирането на числа с практическо значение изисква много повече кюбити, отколкото са налични в близко бъдеще. Друга причина за безпокойство е, че шумът в квантовите вериги може да подкопае резултатите, което изисква допълнителни кюбити за квантова корекция на грешките.


Факторирането на цели числа обикновено се смята за трудно за класически компютър. Но сега се смята, че факторизацията на прости числа може да се осъществи за полиномиално време на квантов компютър. Тази забележителна работа се дължи на Питър У. Шор. За дадено число $n$ той дава алгоритъм на квантов компютър за намиране на реда $r$ на елемент $x$ (mod $n$), вместо да даде алгоритъм на квантов компютър за директно факториране на $n$. Индиректният алгоритъм е осъществим, защото факторизацията може да се сведе до намиране на реда на даден елемент чрез използване на случаен принцип. Сега ще дадем накратко това намаление.

За да се намери фактор на нечетно число n, като се има предвид метод за изчисляване на реда $r$ на $x$, се избира произволно $x (\text{mod} \ n)$, намира се неговият ред $r$ и се изчислява $\text{gcd}(x^{r/2} -1, n)$.

( __GCD__ е съкращение от "Greatest Common Divisor", което на български означава "Най-голям общ делител" (НОД).

В математиката, НОД на две цели числа е най-голямото число, което дели и двете числа без остатък. Например, НОД на числата 12 и 18 е 6, тъй като 6 е най-голямото число, което може да дели както 12, така и 18 без остатък. )

С други думи Алгоритъмът е:

Дадено е нечетно съставно число N, намерете цяло число d, строго между 1 и N, което дели N.

Алгоритъмът на Шор се състои от следните две части:

* Превръщане на задачата за факторизиране в задача за намиране на период. Тази част може да се реализира с класически средства.
* Намиране на периода или квантово намиране на периода, като се използва квантовото преобразуване на Фурие, и е отговорно за квантовото ускорение и използва квантовия паралелизъм.

В алгоритъма на Шор __входното__ е непразно число __N__, а __изходното__ е нетривиален коефициент на __N__

$$\text{INPUT} (N) -> \text{АЛГОРИТЪМ НА ШОР} -> \text{ИЗХОД} (\text{Нетривиален фактор на} N)$$


Алгоритъм: Той съдържа няколко стъпки, като само при стъпка 2 се изисква използването на квантови компютри.

1. Изберете произволно число, да речем $r$, такова, че $r < N$, така че те са взаимно прости помежду си.   Изразът "те са взаимно прости помежду си" означава, че две числа нямат общи делители освен единица. Това означава, че НОД (най-голям общ делител) на тези числа е равен на 1.
2. Използва се квантов компютър, за да се определи неизвестният период $p$ на функцията $f_{r, N} \ (x) = r^x \ \text{mod} \ N$.
3. Ако $p$ е нечетно цяло число, върнете се към стъпка 1. В противен случай преминете към следващата стъпка.
4. Тъй като $p$ е четно цяло число, то $(r^{p/2} - 1)(r^{p/2} + 1) = r^p - 1 = 0 \ \text{mod} \ N$.
5. Сега, ако стойността на $r^{p/2} + 1 = 0 \ \text{mod} N$, върнете се към стъпка 1.
6. Ако стойността на $r^{p/2} + 1 != 0 \  \text{mod} N$, то преминете към следващата стъпка.
7. Изчислете $\text{d} = \text{gcd}(r^{p/2}-1, N)$.
8. Необходимият отговор е __'d'__.

$\underline{\text{Класическа част (Задача за намиране на ред):}}$

Това е класическата част на задачата за намиране на ред. Дадени са $x$ и $N$, такива, че $x<N$ и $\text{gcd}(x, N) = 1$. Редът на $x$ е най-малкото положително цяло число, $y$ такова, че $x^y = 1(\text{mod} \ N)$.

1. Избира се случайно число $n$, такова че $n < N$. Изчислете $\text{gcd}(n, N)$.
2. Това може да се направи с помощта на алгоритъма на Евклид.
3. Ако $gcd(n, N) != 1$, тогава има нетривиален фактор на $N$. Ако $(x+p) = n^{x + p} \ \text{mod}N = n^x \ \text{mod}N = f(x)$.
4. Ако r е нечетно, върнете се към __Стъпка 1__.
5. Ако $n^{p/2} = -1(\text{mod} \ N)$, върнете се към __Стъпка 1__.
6. Числото $\text{gcd}(n^{p/2} +/- 1, \ N)$ е нетривиален фактор на $N$.


$\underline{\text{Квантова част:}}$

Това е квантовата част за намиране на реда. Изберете степен на 2, тогава
$$Q = 2L \ \text{така, че} \ N2 <= Q <= 2N2$$

И разгледайте f = {0, 1, 2, ..., Q-1}

Където $f(y)=1/(Q)^{1/2} \sum_{x=0}^{Q-1}I f(x) I w^{xy}$ и $w = exp(2\pi \ i/Q), \text{i.e.} Q^{th}$ корен на единицата.

* Нека изпълним алгоритъма на Шор с помощта на пример: За да умножим нечетно цяло число N (нека N = 17).
* Изберете цяло число Q, такова че $N^2 <= Q ≤ 2 N^2$ (нека Q = 324).
* След това изберете произволно цяло число n, такова че $\text{gcd}(n, N)=1$, (нека изберем цялото число да бъде n=17).
* След това създайте два квантови регистъра (тези регистри трябва да са заплетени така, че сривът на входния регистър да съответства на срив на изходния регистър)
    * __Входящ регистър__: трябва да съдържа достатъчно кюбити, за да представя числа, големи колкото $Q-1$ (т.е. до 323, така че ни трябват 9 кюбита).
    * __Изходен регистър__: трябва да съдържа достатъчно кюбити, за да представя числа, големи колкото $(N - 1)$ (т.е. до 16, така че ни трябват 4 кюбита).

Сега нека се впуснем в __имплементацията на алгоритъма на Шор__ с помощта на рамката за квантово програмиране ProjectQ. Ще проучим как квантовите принципи като суперпозиция и заплитане ни позволяват да се справим с един от най-предизвикателните проблеми в класическите изчисления: факторизация на цели числа.

```python
import math
import random
import sys
from fractions import Fraction
from builtins import input

def shor(N, attempts = 1):
    
    # build compilation engine list
    rule_set = DecompositionRuleSet(modules=[projectq.libs.math,
                                             projectq.setups.decompositions])
    compilerengines = [AutoReplacer(rule_set),
                       InstructionFilter(high_level_gates),
                       TagRemover(),
                       LocalOptimizer(3),
                       AutoReplacer(rule_set),
                       TagRemover(),
                       LocalOptimizer(3)]

    # make the compiler and run the circuit on the simulator backend
    eng = MainEngine(Simulator(), compilerengines)
    
    for attempt in range(attempts):
        print("-----------------------------------")
        print("\nAttempt #" + str(attempt))
        
        x = random.randint(0, N)
        print("Random x between 0 and N-1 --> ", str(x))
        
        '''If the GCD is not 1, x is a nontrivial factor of N, so we're done'''
        if (math.gcd(x, N) != 1):
            print("\nFactors found classically, re-attempt...")
            continue
        
        '''Otherwise it means that x and N are coprime
        Here the quantum part starts: Shor's algorithm tries to find r, the period of x^a mod n, where n is 
        the number to be factored and x is an integer coprime to n.
        It is important to underline the r  is the smallest positive integer such that x^r = 1 mod N
        '''
        r = find_period(eng, N, x, verbose=False)
        
        '''If r is odd or if x^r/2 = -1 (mod N), choose another x
        EXPLANATION: we already know that x^r/2 is NOT congruent to 1 (mod N), otherwise the order of x would be r/2,
        instead of r. So we have to check only that x^r/2 is NOT congruent to -1 (mod N)
        '''
        if ((r % 2 != 0) or (pow(x, int(r/2), N) == -1)): 
            print("r is odd or x^r/2 = -1 (mod N), re-attempt...")
            continue
            
        print("\nPeriod found: " + str(r))
        
        p = math.gcd(x**int(r / 2) + 1, N)
        q = math.gcd(x**int(r / 2) - 1, N)
        
        if ((not p * q == N) and p * q > 1 and int(1. * N / (p * q)) * p * q == N):
            p, q = p * q, int(N / (p * q))
        if p * q == N and p > 1 and q > 1:
            print("\nFactors found: {} * {} = {}.".format(p, q, N))
            break
        else:
            print("\nBad luck: Found {} and {}".format(p, q))
```

```python
'''Simulating Shor s algorithm at the level of single-qubits gates and CNOTs already takes
quite a big of time for larger numbers than 15.
Returning TRUE allows to use the emulation feature, which does not decompose the modular arithmetic to low-level gates.'''
def high_level_gates(eng, cmd):
    return True
```

```python
def find_period(eng, N, a, verbose=False):
    
    n = int(math.ceil(math.log(N, 2))) # log2(N)

    x = eng.allocate_qureg(n)

    '''Pauli-X gate: acts on a single qubit. It is the quantum equivalent of the NOT gate for classical computers'''
    X | x[0]

    measurements = [0] * (2 * n)  # will hold the 2n measurement results

    ctrl_qubit = eng.allocate_qubit()

    for k in range(2 * n):
        '''Compute a^x mod N conditioned on a control qubit ctrl_qubit in a uniform superposition of 0 and 1'''
        current_a = pow(a, 1 << (2 * n - 1 - k), N) # << is binary left shift, so we are computing a^(2^(2n-1-k)) mod N
        # one iteration of 1-qubit QPE (Quantum Phase Estimation)
        H | ctrl_qubit
        with Control(eng, ctrl_qubit):
            MultiplyByConstantModN(current_a, N) | x

        # perform semi-classical inverse QFT --> Rotations conditioned on previous outcomes
        for i in range(k):
            if measurements[i]:
                R(-math.pi/(1 << (k - i))) | ctrl_qubit
        
        # final Hadamard of the inverse QFT
        H | ctrl_qubit

        # and measure
        Measure | ctrl_qubit
        eng.flush() # flush all gates and execute measurements
        measurements[k] = int(ctrl_qubit)
        if measurements[k]:
            X | ctrl_qubit # reset

        if verbose:
            print("\033[95m{}\033[0m".format(measurements[k]), end="")
            sys.stdout.flush()

    All(Measure) | x # shortcut (instance of) projectq.ops.Tensor
    # turn the measured values into a number in [0,1)
    y = sum([(measurements[2 * n - 1 - i]*1. / (1 << (i + 1)))
             for i in range(2 * n)])

    # continued fraction expansion to get denominator (the period)
    r = Fraction(y).limit_denominator(N-1).denominator

    # return the (potential) period
    return r
```

# Какво е квантова схема

Квантовата схема е квантовият компютърен еквивалент на класическа цифрова схема. Тя е фундаментално понятие в квантовите изчисления, представляващо универсален набор от квантови врати и кюбити, които извършват специфични изчисления.

Пример за структура на квантова схема:

* __Qubits__. Фундаменталните единици на квантовата информация. За разлика от класическите битове (0 или 1), кюбитите съществуват в суперпозиции.

* __Квантови врати__. Подобни на класическите логически врати, но работят с кюбити, използвайки квантови принципи. Примерите включват портата на Хадамард за суперпозиция и портата CNOT за заплитане. Ще обясня CNOT портата по-долу.

* __Заплитане__. Основна характеристика на квантовите вериги, при която състоянията на кюбитите са свързани помежду си, дори на огромни разстояния.

* __Измерване__. Завършвайки квантовите схеми, измерването срива кюбитите от суперпозиция до класически състояния (0 или 1), предоставяйки резултата от изчислението.

__CNOT__ в информатиката, __контролираният NOT гейт__ (също __C-NOT__ или __CNOT__), __контролираният-X гейт__, __контролираният-bit-flip гейт__, __Гейт на Файнман__ или __контролираният Pauli-X__ е квантов логически гейт, който е съществен компонент в конструкцията на квантов компютър, базиран на гейт. Той може да се използва за заплитане и разплитане на състоянията на Бел.Всяка квантова схема може да се симулира с произволна степен на точност, като се използва комбинация от CNOT гейтове и ротации на единични кюбити.
CNOT може да се изрази на базата на Паули по следния начин: 
$$ \text{CNOT} = e^{i \frac{\pi}{4}(I_1 - Z_1)(I_2 - X_2)} = e^{-i \frac{\pi}{4}(I_1 - Z_1)(I_2 - X_2)} $$
CNOT гейтът работи с квантов регистър, състоящ се от 2 кюбита.Вратата CNOT преобръща втория кюбит (целевия кюбит) тогава и само тогава, когато първият кюбит (контролния кюбит) е |1⟩ .Действието на CNOT портата може да бъде представено чрез матрица (форма на матрица на пермутация):

### Ролята на квантовите схеми:

Факторизацията, криптографията, оптимизацията и симулацията на квантови системи са сред трудните проблеми, с които квантовите схеми се справят отлично и които са практически неприложими за конвенционалните компютри.


#### Накратко, квантовите схеми са:

* Градивните елементи на квантовото изчисление
* Използване на уникалните свойства на кюбитите
* Квантови врати
* Заплитане за извършване на изчисления, които са многообещаващи за бъдещето на технологиите

#### Критериите на ДиВинченцо накратко

През 2000 г. физикът Дейвид П. ДиВинченцо създава Критериите на ДиВинченцо като рамка за оценка на практичността и потенциала на квантовата компютърна технология. Тези критерии очертават основните изисквания, необходими за създаването на практичен и мащабируем квантов компютър.

Нека разгледаме тези __стандарти__ по-подробно.

__Мащабируема физическа система__ <br>

Първият критерий изисква наличието на мащабируема платформа за внедряване на квантови битове или кюбити. Това означава да има средства за създаване, манипулиране и надежден контрол на голям брой кюбити. Свръхпроводими вериги и уловени йони са примери за обещаващи кандидати.

__Универсални квантови порти__ <br>

За да може да извършва широк спектър от квантови изчисления, квантовият компютър трябва да разполага с набор от универсални квантови порти, подобни на класическите логически порти.

__Дълги времена на кохерентност на кюбита__ <br>

Квантовата информация е крехка и лесно се влияе от факторите на околната среда. Поради това кюбитите трябва да имат дълго време на кохерентност, което да им позволява да запазват квантовите си състояния за достатъчна продължителност за извършване на изчисления и коригиране на грешки.

__Инициализация и измерване на кюбити с висока точност__ <br>

Точното инициализиране на кюбитите в известно състояние и прецизните резултати от измерванията са от съществено значение за надеждните квантови изчисления. 

__Универсална класическа изчислителна способност__ <br>

Квантовият компютър трябва да може да се свързва безпроблемно с класическите компютри. Това означава ефективна комуникация между квантовите и класическите компоненти за задачи като въвеждане, извеждане и коригиране на грешки.

__Поправка на грешка__ <br>

Квантовите компютри са предразположени към грешки, дължащи се на декохерентност и други фактори. Критериите на DiVincenzo подчертават необходимостта от схема за квантова корекция на грешки, която да осигури надеждни и устойчиви на грешки квантови изчисления.

__Повторими операции с висока степен на достоверност__ <br>

Квантовите операции трябва да могат да се повтарят с висока точност, като се гарантира, че едни и същи изчисления могат да се извършват последователно.

# Да симулираме квантова схема

Целта на симулирането на схема е да се разбере как се държат квантовите операции като суперпозиция и заплитане в контролирана среда без шум. Чрез симулиране на квантови схеми изследователите и разработчиците могат да изследват квантови алгоритми, да тестват техники за квантова корекция на грешки и да правят сравнителен анализ на квантов хардуер, без да се влияят от ограниченията на реалните квантови устройства.

Ще стартираме проста схема, която подготвя 2-кюбитово състояние на Бел и измерва двата кюбита:

```python
simulator = AerSimulator()
# Qiskit Aer simulator backend, we will use it to run ideal (noise free) simulator.
```

```python
# Create circuit
circ = QuantumCircuit(2)  # This creates a quantum circuit with 2 qubits.
circ.h(0)  # Applies a Hadamard gate to qubit 0, putting it into a superposition state.
circ.cx(0, 1)  # Applies a CNOT gate with qubit 0 as the control and qubit 1 as the target. 
circ.measure_all()  # Measures both qubits.

# Transpile for simulator
simulator = AerSimulator()  # Initializes the AerSimulator, a simulator backend in Qiskit.
circ = transpile(circ, simulator)  # Transpiles (or converts) the quantum circuit to be optimized for the simulator backend.

# Run and get counts
result = simulator.run(circ).result()  #  Executes the transpiled quantum circuit on the simulator and retrieves the result.
counts = result.get_counts(circ)  # Extracts the measurement results (counts of each outcome) from the simulation result.

plot_histogram(counts, figsize=(10, 7), title='Bell-State counts')  # Plots a histogram of the measurement outcomes
```

#### Визуализиране на вектора на състоянието на Бел
За да задълбочим разбирането си за квантовото състояние, създадено от нашата схема, сега ще симулираме същата квантова схема без непосредствено измерване и ще визуализираме вектора на състоянието ѝ. Това ще ни позволи да видим пълното квантово състояние и свойствата на заплитане на състоянието на Бел.

```python
# Construct a quantum circuit with 2 qubits
circ = QuantumCircuit(2)

# Apply a Hadamard gate to the first qubit (qubit 0)
# This puts qubit 0 into a superposition state (|0> + |1>) / sqrt(2)
circ.h(0)

# Apply a CNOT gate with qubit 0 as the control and qubit 1 as the target
# This entangles qubit 1 with qubit 0, creating a Bell state
circ.cx(0, 1)

# Save the statevector of the quantum circuit
# The statevector contains the complete information about the quantum state
circ.save_statevector()

# Initialize the AerSimulator with the 'statevector' method
# This method simulates the exact statevector of the quantum system
simulator = AerSimulator(method='statevector')

# Transpile the quantum circuit for the simulator
# Transpilation optimizes the circuit for the specific backend (simulator)
circ = transpile(circ, simulator)

# Run the transpiled circuit on the simulator and get the result
result = simulator.run(circ).result()

# Retrieve the statevector from the result
statevector = result.get_statevector(circ)

# Plot the statevector using a cityscape plot
# The plot shows the real and imaginary components of the statevector amplitudes
plot_state_city(statevector, title='Bell state')
```

#### Запазване на няколко състояния
Можем също така да прилагаме инструкции за запаметяване на няколко места във веригата. Обърнете внимание, че когато правим това, трябва да предоставим уникален етикет за всяка инструкция, за да ги извлечем от резултатите

```python
# Construct quantum circuit without measure
steps = 5
circ = QuantumCircuit(1)
for i in range(steps):
    circ.save_statevector(label=f'psi_{i}')
    circ.rx(i * np.pi / steps, 0)
circ.save_statevector(label=f'psi_{steps}')

# Transpile for simulator
simulator = AerSimulator(method= 'automatic')
circ = transpile(circ, simulator)

# Run and get saved data
result = simulator.run(circ).result()
data = result.data(0)
data
```

#### Можем също така да зададем потребителски вектор на състоянието

Инструкцията set_statevector в Qiskit може да се използва за задаване на потребителски statevector за квантова схема. Входният вектор на състоянието трябва да бъде валидно квантово състояние
($ | \langle \psi | \psi \rangle | = 1 $) .

Нека обясним какво е $ | \langle \psi | \psi \rangle | = 1 $ : <br>

__Квантово състояние и запис на Бра-Кет__
1. Квантово състояние $| \psi \rangle $
    * Квантовото състояние се представя чрез кет $| \psi \rangle $. Този кет е вектор в комплексно векторно пространство, известно като пространство на Хилберт.
2. Bra $ \langle \psi |$
    * Съответният Bra $ \langle \psi |$ е хермитов конюгат (или комплексна конюгирана транспозиция) на кет $| \psi \rangle $.
  
__Вътрешен продукт и нормализация__ <br>

3. Вътрешен продукт $ \langle \psi | \psi \rangle $
    * Изразът $ \langle \psi | \psi \rangle $ представлява вътрешното произведение на квантовото състояние със самото себе си. На математически език това е комплексно число, което описва "припокриването" на състоянието със себе си.
  
4. Условие за нормализация:
    * За да бъде валидно едно квантово състояние, то трябва да бъде нормализирано. Това означава, че общата вероятност на всички възможни изходи трябва да е 1. В контекста на вътрешното произведение това условие се записва по следния начин: $$ | \langle \psi | \psi \rangle | = 1 $$
  
Той потвърждава, че амплитудата на вероятността на състоянието, когато е квадратирана, е равна на 1. Това е от решаващо значение, за да може състоянието да бъде валидно физическо състояние в квантовата механика.

В обобщение, $ | \langle \psi | \psi \rangle | = 1 $ гарантира, че квантовото състояние $| \psi \rangle $ е нормализирано, т.е. има единична норма, което е основно изискване за всяко квантово състояние. Това гарантира, че общата вероятност за всички резултати, описани от състоянието, е равна на 1, като се запазва вероятностната интерпретация на квантовата механика.

Сега нека генерираме случаен вектор на състоянието, да го зададем като начално състояние на нашата квантова схема и да я симулираме.

```python
# Generate a random statevector
num_qubits = 2  # Define the number of qubits
psi = qi.random_statevector(2 ** num_qubits, seed=100)  # Generate a random statevector for 2 qubits

# Create a quantum circuit with the specified number of qubits
circ = QuantumCircuit(num_qubits)

# Set the initial state of the quantum circuit to the generated statevector
circ.set_statevector(psi)

# Save the state of the quantum circuit
circ.save_state()

# Initialize the AerSimulator with the 'statevector' method
simulator = AerSimulator(method='statevector')

# Transpile the quantum circuit for the simulator
circ = transpile(circ, simulator)

# Run the transpiled circuit on the simulator and get the result
result = simulator.run(circ).result()

# Retrieve and print the saved data (statevector) from the result
statevector_data = result.data(0)
print(statevector_data)
```

Възможно е също така да инициализирате симулатора в потребителски вектор на състоянието, като използвате инструкцията __initialize__. За разлика от инструкцията set_statevector тази инструкция се поддържа и от реалните устройства, като се разгръща към инструкциите reset и standard gate.

```python
# Use initilize instruction to set initial state
circ = QuantumCircuit(num_qubits)
circ.initialize(psi, range(num_qubits))
circ.save_state()

# Transpile for simulator
simulator = AerSimulator(method= 'statevector')
circ = transpile(circ, simulator)

# Run and get result data
result = simulator.run(circ).result()
result.data(0)
```
